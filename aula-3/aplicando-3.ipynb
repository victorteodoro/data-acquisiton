## Questão 7

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 
```

```python
# URL da home do Mackenzie
url_mack = "https://www.mackenzie.br"

# Extração do html da pagina do Mackenzie
html_mack = requests.get(url_mack).text

# Parser do HTML para o objeto BeautifulSoup
soup_mack = BeautifulSoup(html_mack, "html.parser")

# Acesso aos elementos da pagina

# Pagina completa
print(soup_mack.prettify())
```

```python
mack_media_urls = []
for link in soup_mack.find(id="c4602").find_all("a"):
    mack_media_urls.append(
        {
            "url": link.get("href"),
            "title": link.get("title")
        }
    )
```

```python
mack_media_df = pd.DataFrame(mack_media_urls)
```

```python
mack_media_df.shape
```
