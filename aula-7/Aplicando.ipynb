```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
july_prices = pd.read_csv('precos_glp_07.csv', sep=';')
july_prices.head()
```

```python
aug_prices = pd.read_csv('precos_glp_08.csv', sep=';', header=None, names=july_prices.columns)
aug_prices.head()
```

```python
july_prices.sort_values(by='Data da Coleta', ascending=True, inplace=True)

july_prices = july_prices.drop_duplicates(subset=['CNPJ da Revenda'], keep='last')
```

```python
aug_prices.sort_values(by='Data da Coleta', ascending=True, inplace=True)

aug_prices = july_prices.drop_duplicates(subset=['CNPJ da Revenda'], keep='last')
```

```python
july_prices.loc[july_prices['CNPJ da Revenda'] == '19.791.896/0140-71']
```

```python
prices = pd.merge(july_prices, aug_prices[['CNPJ da Revenda', 'Valor de Venda']], how='inner', on='CNPJ da Revenda', suffixes=[' Jul', ' Aug'])
prices.head()
```

```python
prices['Valor de Venda Jul'] = prices['Valor de Venda Jul'].str.replace(',', '.')
prices['Valor de Venda Jul'].astype(float).mean()
```

```python
prices['Valor de Venda Aug'] = prices['Valor de Venda Aug'].str.replace(',', '.')
prices['Valor de Venda Aug'].astype(float).mean()
```

```python
prices['Valor de Venda Jul'] = prices['Valor de Venda Jul'].astype(float)
prices['Valor de Venda Aug'] = prices['Valor de Venda Aug'].astype(float)

prices['Valor de Venda Jul'].loc[prices['Regiao - Sigla'] == 'CO'].mean()
```

```python
prices['Valor de Venda Jul'].loc[prices['Bandeira'] == 'MINASGAS'].mean()
```

```python
(prices['Valor de Venda Jul'].mean() - prices['Valor de Venda Aug'].mean()) / prices['Valor de Venda Jul'].mean()
```

```python
cost_of_living = pd.read_csv('costofliving.csv')
cost_of_living.head()
```

```python
cost_of_living = cost_of_living.select_dtypes(include=['number'])
cost_of_living = cost_of_living.drop(columns=['Rank'])
cost_of_living.head()
```
